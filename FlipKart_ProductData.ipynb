```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
def specification_of_product(url):
    
    
    
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    specifications=soup.find('div',class_="_3dtsli")
    
    for each in specifications:
        c1=each.find_all('td',attrs={'class':'_1hKmbr'})
        
        c2=each.find_all('li',attrs={'class':'_21lJbe'})
        
    s1=c1[0].text+'->'+c2[0].text
    s2=c1[1].text+'->'+c2[1].text
    s3=c1[2].text+'->'+c2[2].text
    s4=c1[3].text+'->'+c2[3].text
    
   
    
    return s1,s2,s3,s4
```

```python
def offers_product(url):
    r=requests.get(url)
    combo_offers = []
    soup=BeautifulSoup(r.text,'html.parser')
    offers=soup.find('div',class_="_3TT44I")
    for combo_offer_element in offers:
        combo_offer = offers.text.strip()
        combo_offers.append(combo_offer)
        
    return combo_offers


```

```python
def scrape_flipkart_product(df,url):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    
 
    
    title = soup.find('span', {'class': "B_NuCI"}).text.strip()
    price = soup.find('div', {'class': "_30jeq3 _16Jk6d"}).text.strip().replace('₹', '')
    stars=soup.find('div',class_="col-12-12 _1azcI6").text.replace('★','')
    ratings_reviews=soup.find('span',class_="_2_R_DZ").text.replace('\xa0&\xa0',' and ')
    specifications=specification_of_product(url)
    offers_available=offers_product(url)
    
        
    data={'Title':[title],'Price':[price],'Stars':[stars],'Ratings and Reviews':[ratings_reviews],'Top Specifications':[specifications],
          'Offers Available':offers_available}
    new_df=pd.DataFrame(data)
    
    df=df.append(new_df,ignore_index=True)
    
    return df
```
